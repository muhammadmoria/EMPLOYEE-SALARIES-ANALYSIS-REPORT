```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("Employee_Salaries.csv")
```

```python
data.head(10)
```

```python
purple = ['#d5a2d5', '#b681b6', '#9e60a1', '#8d4690', '#7d007d']
```

```python
plt.figure(figsize=(8, 6))
data['Gender'].value_counts().plot(kind='bar', color=purple)
plt.title('Gender Distribution of Employees')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 14))
data.groupby('Department_Name')['Base_Salary'].mean().sort_values().plot(kind='barh', color=purple)
plt.title('Average Salary by Department')
plt.xlabel('Average Salary')
plt.ylabel('Department')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Overtime_Pay'], bins=30, color=purple[2])
plt.title('Distribution of Overtime Pay')
plt.xlabel('Overtime Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Longevity_Pay'], bins=30, color=purple[3])
plt.title('Distribution of Longevity Pay')
plt.xlabel('Longevity Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Base_Salary'], data['Overtime_Pay'], color=purple[4], alpha=0.5)
plt.title('Base Salary vs. Overtime Pay')
plt.xlabel('Base Salary')
plt.ylabel('Overtime Pay')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='Base_Salary', by='Gender', grid=False, color=dict(boxes=purple[2], whiskers=purple[1], medians=purple[4], caps=purple[3]))
plt.title('Base Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Base Salary')
plt.suptitle('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data['Gender'].value_counts().plot(kind='pie', autopct='%1.2f%%', colors=purple)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
sns.pairplot(data[['Base_Salary', 'Overtime_Pay', 'Longevity_Pay']], diag_kind='kde', plot_kws={'color': purple[2]})
plt.show()
```
